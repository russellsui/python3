```python
import matplotlib as mpl
import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')
fig = plt.figure()
ax = plt.axes()
```

```python
import numpy as np
mean = [0,0]
cov = [[1,1],[1,2]]
x, y = np.random.multivariate_normal(mean,cov,10000).T
plt.hexbin(x,y,gridsize = 30, cmap = 'Blues')
cb = plt.colorbar(label = 'count in bin')
```

```python
fig = plt.figure()
ax = plt.axes()
x = np.linspace(0,10,1000)
ax.plot(x, np.sin(x))
```

```python
# make two axes
fig,ax = plt.subplots(2)
ax[0].plot(x,np.sin(x))
ax[1].plot(x,np.cos(x))
```

```python
# change the color and style
plt.plot(x, np.sin(x - 0), color = 'blue')
plt.plot(x, np.sin(x - 1), color = 'g')
plt.plot(x, np.sin(x - 2), color = '0.75')
# use RGB tuple
plt.plot(x, np.sin(x - 3), color = (1.0,0.2,0.3))
plt.plot(x, np.sin(x - 4), color = 'chartreuse')
```

```python
# could also use linestyle to change the style
plt.plot(x, x + 0, linestyle = 'solid', color = 'blue')
plt.plot(x, x + 1, linestyle = 'dashed', color = (0.2,0.8,0.35))
plt.plot(x, x + 2, linestyle = 'dashdot', color = 'r')
plt.plot(x, x + 3, ':y')
```

```python
# could also change the limit of axes
plt.plot(x, np.sin(x))
plt.xlim = (-10,10)
plt.ylim = (-2,2)
```

```python
plt.plot(x, np.sin(x))
plt.axis([-5,5,-0.5,0.5])
```

```python
# set the title and label for the plot
plt.plot(x,np.sin(x))
plt.title('y = sin(x)')
plt.xlabel('x')
plt.ylabel('y')
```
