```python
# the most general case to create a bar chart is plotting a bar chart by sending
# the x component and the height of the bar
%matplotlib notebook
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
# create the linear_data using the array and then set is as the height of our bars
linear_data = np.array([1,2,3,4,5,6,7,8])
# create a list of values to be the xaxis of the figure
xvals = range(len(linear_data))
# plt.bar(xvalues, height of the bars, width)
plt.bar(xvals, linear_data, width = 0.3)
```

```python
plt.xlabel('value')
plt.title('My Bar chart')
```

```python
# create another bar chart next to the original one
new_vals = []
quadratic_data = linear_data ** 2
# Notice that the center of the new bar is at x+0.3, since the width of the original bar was =.3
# so the range of the original bar is x-0.15 ~ x+0.15, the range of the new bar is 
# x + 0.15 ~ x + 0.45, there is no overlap of bars at all
for item in xvals:
    new_vals.append(item + 0.3)
plt.bar(new_vals, quadratic_data, width = 0.3)
```

```python
# actually bar charts is the figure that prof does not really want to plot
# create error terms for bar charts
from random import randint
# create a random error list. For each element in this list, it is an integer from 0 to 15
linear_err = [randint(0,15) for x in range(len(linear_data))]
# replot the bar charts, the parameters are same as before except for an additive error terms
# on y axis, pass the error list to the yerr para 
# and then on the figure there will be a black line on the y direction that denotes the error part
plt.bar(xvals, linear_data, width = 0.3, yerr = linear_err)
```

```python
plt.figure()
# we could also stack the two bars together
xvals = range(len(linear_data))
# first plot the regular bar plot
plt.bar(xvals, linear_data, width = 0.3, color = 'blue')
# then plot the second bar plot at the exactly same location with the linear ones
# add the 'bottom = ?' parameter to indicate that the new bar chart is covered on the first one
plt.bar(xvals, quadratic_data, width = 0.3, bottom = linear_data, color = 'red')
```

```python
plt.figure()
# we could also make a bar chart horizontally
xvals = range(len(xvals))
# In this case, we should use barh instead of bar, and use height instead of width
plt.barh(xvals, linear_data, height = 0.3, color = 'orange')
# if we also wanna make stacks of figures, use left instead of bottom
linear_error = [randint(0,15) * 0.1 for item in range(len(linear_data))]
# we could also make some errors , remember to use xerr instead of yerr here
plt.barh(xvals, quadratic_data, height = 0.3, left = linear_data, color = 'green', xerr = linear_error)
```
