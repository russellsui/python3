```python
import matplotlib as mpl
import matplotlib.pyplot as plt
# %matplotlib inline magic is not interactive
%matplotlib notebook
plt.plot(3,2,'.')
```

```python
from matplotlib.backends.backend_agg import FigureCanvasAgg
from matplotlib.figure import Figure

fig = Figure()
canvas = FigureCanvasAgg(fig)

ax = fig.add_subplot(111)
ax.plot(3,2,'.')
canvas.print_png('test.png')

```

```python
plt.figure()
plt.plot(3,2,'o')
#plt.gca() is getting the current axes
ax = plt.gca()
# setting the xlim =[0,6], ylim = [0,10]
ax.axis([0,6,0,10])
```

```python
# could plot several points in a same figure
plt.figure()
plt.plot(1.5, 1.5, 'o')
plt.plot(2, 2, 'o')
plt.plot(2.5, 2.5, 'o')
```
