```python
# first create a 2*2 subplots, here we intentionally do not share y axis
%matplotlib notebook
import numpy as np
import matplotlib.pyplot as plt
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, sharex = True)
axes = [ax1, ax2, ax3, ax4]
for n in range(0, len(axes)):
    sample_size = 10 ** (n + 1)
    sample = np.random.normal(0, 1, size = sample_size)
    axes[n].hist(sample)
    axes[n].set_title('n = {}'.format(sample_size))
```

```python
# By default, the bin size is 10, but now we set the bin size to be 100
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, sharex = True)
axes = [ax1, ax2, ax3, ax4]
for n in range(0, len(axes)):
    sample_size = 10 ** (n + 1)
    sample = np.random.normal(0, 1, size = sample_size)
    axes[n].hist(sample, bins = 100)
    axes[n].set_title('n = {}'.format(sample_size))
# using plt.hist(sample, bin = ?) to resize the bin
```

```python
plt.figure()
Y = np.random.normal(0, 1, size = 10000)
# create 10000 points between 0 and 1
X = np.random.random(size = 10000)
plt.scatter(X, Y)
```

```python
# use grid spec subplot
import matplotlib.gridspec as gridspec
plt.figure()
gspec = gridspec.GridSpec(3,3)
# it's like using our own way to create subplots, not seperating a big axes equally but
# by grid
top_histogram = plt.subplot(gspec[0, 1:])
side_histogram = plt.subplot(gspec[1:, 0])
lower_right = plt.subplot(gspec[1:, 1:])
```

```python
Y = np.random.normal(0, 1, size = 10000)
X = np.random.random(size = 10000)
# lower_right subplot plot the scatter plot of X and Y
lower_right.scatter(X, Y)
# The top subplot make a histogram of the random points
top_histogram.hist(X, bins = 100)
s = side_histogram.hist(Y, bins = 100, orientation = 'horizontal')
# æ€»ä¹‹å¯ä»¥ç”¨axesæ¥ç”»å›¾ï¼Œä¹Ÿå¯ä»¥ç”¨gridspecï¼Œä¹Ÿå¯ä»¥ç”¨subplot,è¿è¡Œäº†ä¸‰æ¬¡æ‰€ä»¥æœ‰ä¸‰ä¸ªå›¾
```

```python
# we could use clear to clear the figure
top_histogram.clear()
top_histogram.hist(X, bins = 100, normed = True)
side_histogram.clear()
#'normed = True' statement let the scale of frequency changed to between 0 and 1
# invert_xaxis() let the x axis shows on the right side
side_histogram.hist(Y, bins = 100, orientation = 'horizontal', normed = True)

side_histogram.invert_xaxis()
```

```python
# we could also use set_xlim, set_ylim to set the limit of the figure
for ax in [top_histogram, lower_right]:
    ax.set_xlim(0,1)
for ax in [side_histogram, lower_right]:
    ax.set_ylim(-5,5)
```
