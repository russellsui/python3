```python
%matplotlib notebook
import matplotlib.animation as animation
import numpy as np
import matplotlib.pyplot as plt
n = 100
# returns n normal values
x = np.random.randn(n)

def update(curr):
    
# first of all we need to check if curr == n,i.e. the process ends
    if curr == n:
        a.event_source.stop()
    plt.cla()
# create bins with some space between using arange
    bins = np.arange(-4, 4, 0.5)
    plt.hist(x[:curr], bins = bins)
    plt.axis([-4,4,0,30])
    plt.gca().set_title('Sampling the normal distribution')
    plt.gca().set_ylabel('Frequency')
    plt.gca().set_xlabel('Value')
# plt.annotate(text, a list [x,y] to specify the location)
    plt.annotate('n = {}'.format(curr), [3,27])
    
```
