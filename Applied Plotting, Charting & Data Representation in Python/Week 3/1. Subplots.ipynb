```python
%matplotlib notebook
import numpy as np
import matplotlib.pyplot as plt
#plt.subplot(nrows,ncols,index)
plt.figure()
plt.subplot(1,2,1)
linear_data = np.arange(8)
plt.plot(linear_data, '-o')
```

```python
# Now we specify the second subplot,it is on the second index position
# see? If we plot the second one, it shows next to the first plot
quadratic_data = linear_data ** 2
plt.subplot(1,2,2)
plt.plot(quadratic_data,'-o')
```

```python
# we could also stack the quuadratic plot on the first subplot,其实就是相当于切换坐标轴的意思
plt.subplot(1,2,1)
plt.plot(quadratic_data, '-o')
```

```python
# But when we put the quadratic data on the first subplot, we could notice that the axis of the first subplot has
# changed coercely to be the same as the second subplot
plt.figure()
ax1 = plt.subplot(1,2,1)
plt.plot(linear_data, '-o')
ax2 = plt.subplot(1,2,2,sharey = ax1)
plt.plot(quadratic_data, '-x')
```

```python
# And we could draw even more plots in a figure
# use plt.subplots to cut it into 9 axises
fig, ((ax1, ax2, ax3), (ax4, ax5, ax6), (ax7, ax8, ax9)) = plt.subplots(3,3, sharex = True, sharey = True)
ax5.plot(linear_data, '-o')
```

```python
# we could also set the labels of each axes visible, first loop through each axes
for ax in plt.gcf().get_axes():
    for label in ax.get_xticklabels() + ax.get_yticklabels():
        label.set_visible(True)
plt.gcf().canvas.draw()
```
