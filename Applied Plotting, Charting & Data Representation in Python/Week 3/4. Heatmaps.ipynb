```python
# Heatmaps is a way of visualizing 3-D data
# 比如在2-D世界地图上给一个降水量数据，用来表示热度，这便是一个可以用heatmaps来visualize的3-D 数据
import matplotlib.pyplot as plt
import numpy as np
plt.figure()
Y = np.random.normal(0, 1, size =10000)
X = np.random.random(size = 10000)
# we could first create a 2-D histogram using hist2d 
_ = plt.hist2d(X, Y, bins = 25)
```

```python
# change the bin sizes
plt.figure()
# bins 是一行或者一列有多少组，不是一共有多少组
_ = plt.hist2d(X, Y, bins =100)
```

```python
# could use colorbar() as a legend for the hist2d
plt.colorbar()
```
