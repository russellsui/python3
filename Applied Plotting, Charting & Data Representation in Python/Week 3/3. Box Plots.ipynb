```python
%matplotlib notebook
import numpy as np
import matplotlib as mpl
import pandas as pd
normal_sample = np.random.normal(0, 1, size = 10000)
random_sample = np.random.random(size = 10000)
gamma_sample = np.random.gamma(2, size = 10000)

df = pd.DataFrame({'normal': normal_sample, 'random': random_sample, 'gamma': gamma_sample})
# we could use df.describe to get some summary statistics of the sample, returns a dataframe of 
# summary statistics
df.describe()
```

```python
import matplotlib.pyplot as plt
plt.figure()
# whis ='range' will get rid of the scatter point on the plot and gives back a straight line
# whisker values 须子点？
_ = plt.boxplot(df['normal'], whis = 'range')
```

```python
# plt.clf() will clear the current figure on the axes
plt.clf()
# set the parameter as a list of dataframes to plot three boxplots in a same figure
_ = plt.boxplot([df['normal'], df['random'],df['gamma']], whis = 'range')
```

```python
# now we plot a new figure to better check their behaviour on a histogram
plt.figure()
# recall that we could use df['colname'].tolist() to convert the values within a series to a list
_ = plt.hist(df['gamma'].tolist(), bins = 100)
```

```python
# we could also overlay one box on the top of another within a figure
import mpl_toolkits.axes_grid1.inset_locator as mpl_il

plt.figure()
plt.boxplot([df['normal'], df['random'], df['gamma']], whis = 'range')
# use mpl_il.inset_axes to set a new axes within the current axes, use this function to locate the new axes
ax2 = mpl_il.inset_axes(plt.gca(), width = '60%', height = '40%', loc = 2)
ax2.hist(df['gamma'], bins =100)
ax2.margins(x = 0.5)
```

```python
# we could also use yaxis.tick_right() to move the tick of the subaxis to the right of the figure
ax2.yaxis.tick_right()
```
